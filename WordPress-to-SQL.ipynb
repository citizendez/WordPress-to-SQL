```python
import pandas as pd
import xmltodict, json
```

```python
#reading in xml file getting ordered dictionary
with open('Data/80portal.wordpress.2020-12-17-CLIENTS.xml') as xml_file:
    data_dict = xmltodict.parse(xml_file.read())
xml_file.close()
```

```python
#convert ordered dictionary to json
od_data = json.dumps(data_dict, indent=4, sort_keys=True)
print(od_data)
```

```python
#loading ordered dictionary to json
json_data = json.loads(od_data)
json_data
```

```python
#create data frame from json
items = json_data['rss']['channel']['item']
items
df = pd.DataFrame(items)
df
```

```python
#initiate new data frame
df_clients = pd.DataFrame()
```

```python
df_clients = pd.DataFrame()
for i, row in df.iterrows():
    dic_lst = row['wp:postmeta']
    name = ''
    contact = ''
    email = '' 
    post_id = row['wp:post_id']
    title = row['title']
    #print(f'{meta_key} ')
    #print(f'{meta_value}')
    #print(f'____________')
    
    for dic in dic_lst:
        meta_key = dic['wp:meta_key']
        meta_value = dic['wp:meta_value']
        if meta_key == 'name': 
            name = meta_value
        elif meta_key == 'contact': 
            contact = meta_value
        elif meta_key == 'email': 
            email = meta_value
        else:
            pass
    client_dict = {'poc': name, 'contact_info': contact, 'email': email, 'client_id': post_id, 'title': title}
    df_clients = df_clients.append(client_dict, ignore_index=True)
#print(f'{meta_key} ')
#print(f'{meta_value}')
#print(f'____________')
df_clients.set_index('client_id', inplace=True)
df_clients = pd.DataFrame(df_clients, columns = ['client','poc', 'contact_info', 'email'])
df_clients.to_csv('data/clients.csv')
df_clients
```
